```python
import pandas as pd

%precision %.2f
pd.options.display.float_format = '{:,.2f}'.format
```

```python
ano = "2018"
fonte = f"MICRODADOS_ENEM_{ano}.csv"

dados = pd.read_csv(fonte, encoding="ISO-8859-1", delimiter=";")
dados.shape
```

```python
dados.head()
```

```python
dados.columns.values
```

```python
colunas_de_notas = ['NU_NOTA_CN', 'NU_NOTA_CH', 'NU_NOTA_LC',
       'NU_NOTA_MT', 'NU_NOTA_COMP1',
       'NU_NOTA_COMP2', 'NU_NOTA_COMP3', 'NU_NOTA_COMP4', 'NU_NOTA_COMP5',
       'NU_NOTA_REDACAO']
dados[colunas_de_notas].describe()
```

```python
import numpy as np

seed = 43278
np.random.seed(seed)

amostra = dados.sample(frac = 0.025)
amostra.shape
```

```python
amostra.head()
```

```python
amostra[colunas_de_notas].describe()
```

```python
destino = f"MICRODADOS_ENEM_{ano}_SAMPLE_{seed}.csv"

amostra.to_csv(destino, index=False)
```
